```python
import pandas as pd
df = pd.read_csv('sales_records.csv')
df
```

```python
#df2 = df.groupby(['country'])['amount'].sum()
df2 = df.groupby(['country', 'product']).agg({'amount':"sum", 'quantity': "sum", "start_date": 'min',"end_date": 'max'})
df3 = df2.groupby(['country'])['amount'].sum()
```

```python
df3.plot(kind='bar', figsize=(20,8), color="red", xlabel = 'PAÍSES', ylabel = "GANANCIA")
df3
```

```python
df4 = df2.groupby(['country'])['quantity'].sum()
df4.plot(kind='bar', figsize=(20,8), color="green", xlabel = 'PAÍSES', ylabel = "CANTIDAD VENDIDA")
df3
```
